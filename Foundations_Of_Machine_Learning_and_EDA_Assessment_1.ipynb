Question 1 : What is the difference between AI, ML, DL, and Data Science? Provide a brief explanation of each.

AI is the broadest concept, focusing on developing systems that mimic human intelligence, while ML, DL, and Data Science are specialized subfields with distinct roles. ML is a subset of AI for pattern recognition and predictive modeling, DL is a subset of ML using neural networks for complex learning, and Data Science is an interdisciplinary field dedicated to extracting insights and knowledge from data using various analytic and modeling tools.​

Artificial Intelligence (AI)

AI refers to the development of machines and systems that can imitate human decision-making and reasoning. It encompasses all techniques that allow computers to simulate intelligent behavior, which includes logic-based systems, rule-based systems, and learning systems.​

Machine Learning (ML)

ML is a subset of AI that uses statistical and algorithmic approaches to allow computer systems to learn patterns from data and improve over time without explicit programming. Common ML techniques include regression, classification, and clustering.​

Deep Learning (DL)

DL is a specialized subset of ML centered on artificial neural networks with multiple layers, enabling modeling of highly complex patterns and large-scale data processing. Typical DL applications include image recognition, natural language understanding, and speech-to-text systems.​

Data Science

Data Science is the interdisciplinary field focused on collecting, processing, analyzing, and extracting insights from structured and unstructured data. It combines methods from statistics, computer science, domain expertise, and visualization, and often leverages ML and sometimes DL to construct models for business analytics, predictions, and decision-making.

Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them?

Overfitting occurs when a model learns the training data too closely—including noise—resulting in high training accuracy but poor test accuracy. Underfitting is when a model is too simple to capture data patterns, leading to poor performance on both training and test sets.​

Detection
- Overfitting: High accuracy on training data, low accuracy on validation/test data, or large gaps between training vs. test performance.​

- Underfitting: Poor accuracy on both training and test data; the model fails to learn underlying patterns.​

Prevention (Short List)
- Overfitting: Use cross-validation, regularization techniques (L1/L2), simpler models, early stopping, and increase training data.​

- Underfitting: Use more complex models, better features, longer training, and reduce excessive regularization.

Question 3: How would you handle missing values in a dataset? Explain at least three methods with examples.

Three common ways to handle missing values are deletion, imputation, and using advanced algorithms. Each approach suits different data types and problem scenarios.​

1. Deletion
Remove rows or columns with missing values.
Example: Dropping rows where 'age' is missing from a data frame using df.dropna(subset=['age']) in Pandas.​

2. Imputation
Fill missing values with calculated estimates, like mean, median, or mode.
Example: Replace missing 'salary' with the average using df['salary'].fillna(df['salary'].mean()) in Pandas.​

3. Advanced Imputation Algorithms
Predict missing values using algorithms like regression, k-nearest neighbors (KNN) imputation, or multiple imputation.
Example: Use KNN to impute missing 'height' values by calculating based on similar data points.​

These techniques are chosen based on the amount, type, and pattern of missingness in your dataset.


Question 4: What is an imbalanced dataset? Describe two techniques to handle it (theoretical + practical).

An imbalanced dataset has a disproportionate number of samples in one class compared to others, meaning the majority class dominates and the minority class is underrepresented. This leads to models biased toward the majority class and poor performance on the minority class, which is often the focus (e.g., fraud detection or rare disease prediction).​

1. Resampling (Theoretical + Practical)
Theory: Balance data by increasing minority class samples (oversampling) or decreasing majority class samples (undersampling).

Practical Example: Use the SMOTE algorithm in Python’s imblearn library to oversample minority samples:
from imblearn.over_sampling import SMOTE
X_res, y_res = SMOTE().fit_resample(X, y).​

2. Class Weighting (Theoretical + Practical)
Theory: Assign higher penalty to misclassifications of the minority class during model training, guiding the model to focus more on minority instances.

Practical Example: In scikit-learn’s LogisticRegression, use class_weight='balanced' to automatically weight classes by their frequency:
model = LogisticRegression(class_weight='balanced').​

These methods help ensure that models are not biased toward predicting only the majority class and improve classification of minority samples.


Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.

Feature scaling is vital in machine learning because it ensures all features contribute equally to model learning, improves accuracy, speeds up convergence for gradient-based algorithms, and makes distance-based models (like KNN and SVM) work correctly.​

Min-Max Scaling

- Definition: Rescales features to a fixed range, usually.​

- Use case: Best for features needing a strict upper/lower bound or when data is uniformly distributed.

- Example: sklearn.preprocessing.MinMaxScaler() in Python.​


Standardization

- Definition: Transforms data to have zero mean and unit variance.

- Use case: Useful for features with Gaussian distributions or algorithms assuming centered data (like PCA, Logistic Regression).

- Example: sklearn.preprocessing.StandardScaler() in Python.

Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?

Label Encoding assigns each category a unique integer, which can imply order, while One-Hot Encoding creates a separate binary column for each category, representing them without implied ranking.​

Label Encoding

- How it works: Converts categories to integers (e.g., "Red" = 0, "Blue" = 1, "Green" = 2).

- Best for: Ordinal data (where order matters, e.g., "Low," "Medium," "High").

- Pros: Memory efficient, does not increase dimensionality.

- Cons: Imposes artificial numerical order for nominal data, which can mislead certain ML algorithms.​​

One-Hot Encoding

- How it works: Creates new columns for each category, filling with 1 (category present) or 0 (not present).

- Best for: Nominal data (no inherent order, e.g., "Country," "Color").

- Pros: No risk of misinterpreting order; preserves pure categorical meaning.

- Cons: Increases feature space, may be resource-intensive for many categories.​​

When to Prefer Each

- Label Encoding: Use for ordinal features—where categories have a clear order and the value semantics are meaningful.​

- One-Hot Encoding: Use for nominal features—where categories are unranked, or with ML models sensitive to feature order, like linear regression or neural networks

```python
import pandas as pd

# Load the dataset
try:
    df = pd.read_csv('googleplaystore.csv')

    # Clean the data as in the original cell
    df.dropna(subset=['Rating'], inplace=True)
    df = df[df['Category'] != '1.9']

except FileNotFoundError:
    print("Error: 'googleplaystore.csv' not found. Please upload the file or provide the correct path.")
    df = None # Set df to None if file not found

# Display info about the cleaned dataframe (optional, but helpful)
if df is not None:
    print("DataFrame loaded and cleaned successfully.")
    df.info()
```

```python
df.dropna(subset=['Rating'], inplace=True)
df = df[df['Category'] != '1.9']
print("\nMissing values after cleaning:")
print(df[['Rating', 'Category']].isnull().sum())
```

```python
average_rating_by_category = df.groupby('Category')['Rating'].mean()
display(average_rating_by_category)
```

```python
highest_rated_category = average_rating_by_category.idxmax()
lowest_rated_category = average_rating_by_category.idxmin()

print(f"Category with the highest average rating: {highest_rated_category} ({average_rating_by_category[highest_rated_category]:.2f})")
print(f"Category with the lowest average rating: {lowest_rated_category} ({average_rating_by_category[lowest_rated_category]:.2f})")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x=average_rating_by_category.index, y=average_rating_by_category.values)
plt.xticks(rotation=90)
plt.title("Average App Rating by Category")
plt.xlabel("Category")
plt.ylabel("Average Rating")
plt.tight_layout()
plt.show()
```

Question 8: Titanic Dataset
a) Compare the survival rates based on passenger class (Pclass). Which class had the highest survival rate, and why do you think that happened?
b) Analyze how age (Age) affected survival. Group passengers into children (Age < 18) and adults (Age ≥ 18). Did children have a better chance of survival?


```python
import pandas as pd

# Read the titanic.csv file into a pandas DataFrame
df_titanic = pd.read_csv('/content/titanic.csv')

# Display the first few rows of the DataFrame
display(df_titanic.head())

# Display information about the DataFrame
display(df_titanic.info())
```

```python
# Calculate the survival rate for each passenger class
survival_rate_by_pclass = df_titanic.groupby('Pclass')['Survived'].mean()

# Print the survival rates
print("Survival Rate by Passenger Class:")
display(survival_rate_by_pclass)

# Identify the class with the highest survival rate
highest_survival_class = survival_rate_by_pclass.idxmax()
print(f"\nPassenger class with the highest survival rate: Pclass {highest_survival_class} ({survival_rate_by_pclass[highest_survival_class]:.2f})")

# Create a bar plot for visualization
plt.figure(figsize=(8, 5))
sns.barplot(x=survival_rate_by_pclass.index, y=survival_rate_by_pclass.values)
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class (Pclass)")
plt.ylabel("Survival Rate")
plt.xticks(ticks=[0, 1, 2], labels=[1, 2, 3]) # Assuming Pclass values are 1, 2, 3
plt.tight_layout()
plt.show()
```

```python
# Create 'AgeGroup' column
df_titanic['AgeGroup'] = df_titanic['Age'].apply(lambda x: 'Child' if x < 18 else ('Adult' if pd.notnull(x) else 'Unknown'))

# Calculate survival rates by age group
survival_rate_by_age_group = df_titanic.groupby('AgeGroup')['Survived'].mean()

# Print the survival rates by age group
print("Survival Rate by Age Group:")
display(survival_rate_by_age_group)

# Create a bar plot to visualize the survival rates by age group
plt.figure(figsize=(8, 5))
sns.barplot(x=survival_rate_by_age_group.index, y=survival_rate_by_age_group.values)
plt.title("Survival Rate by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Survival Rate")
plt.tight_layout()
plt.show()
```

```python
print("## Summary of Findings")
print("")
print("Based on the analysis of the Titanic dataset:")
print("")
print(f"- **Survival Rate by Passenger Class:**")
print(f"  - First Class (Pclass 1): {survival_rate_by_pclass[1]:.2%} survival rate")
print(f"  - Second Class (Pclass 2): {survival_rate_by_pclass[2]:.2%} survival rate")
print(f"  - Third Class (Pclass 3): {survival_rate_by_pclass[3]:.2%} survival rate")
print("")
print(f"  The passenger class with the highest survival rate was **First Class (Pclass 1)**.")
print("")
print(f"- **Survival Rate by Age Group:**")
print(f"  - Children (< 18 years): {survival_rate_by_age_group['Child']:.2%} survival rate")
print(f"  - Adults (>= 18 years): {survival_rate_by_age_group['Adult']:.2%} survival rate")
print("")
print(f"  Children had a **better chance of survival** compared to adults, with a survival rate of {survival_rate_by_age_group['Child']:.2%} for children versus {survival_rate_by_age_group['Adult']:.2%} for adults.")
print("")
print("The bar plots created visually support these findings, clearly showing the higher survival rates for First Class passengers and for children.")
```
